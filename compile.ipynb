```python
dir = "./"

import os

# Iterate files in dir
for filename in os.listdir(dir):

    # If filename contains Taylor-swift-
    if "Taylor-swift-" in filename:
        # Remove Taylor-swift- from the filename
        new_filename = filename.replace("Taylor-swift-", "")
        
        # Overwrite the file with the new filename
        os.rename(dir + filename, dir + new_filename)
```
